```python
import matplotlib.pyplot as plt

def plot_values(time):
    entries = [1000, 10000, 100000, 1000000, 10000000]
    
    plt.figure(figsize=(10, 6))
    
    # Configuração do fundo branco
    plt.gcf().set_facecolor('white')  # Fundo branco para a figura
    
    # Plotando a primeira curva
    plt.plot(entries, time[0], marker='o', color='skyblue', linestyle='-', label='Curva utilizando Insertion Sort')
    
    # Plotando a segunda curva
    plt.plot(entries, time[1], marker='o', color='orange', linestyle='-', label='Curva utilizando Heap')
    
    plt.xscale('log')  # Escala logarítmica para o eixo X
    plt.yscale('log')  # Escala logarítmica para o eixo Y
    plt.xlabel('Tamanho das Entradas')
    plt.ylabel('Tempo (s)')
    plt.title('Gráfico de Execução dos Algoritmos', fontweight='bold')
    
    # Adicionando tracejados nos pontos
    for i in range(len(entries)):
        plt.axvline(x=entries[i], color='gray', linestyle='--', lw=0.7, alpha=0.7)
        plt.axhline(y=time[0][i], color='skyblue', linestyle='--', lw=0.7, alpha=0.7)
        plt.axhline(y=time[1][i], color='orange', linestyle='--', lw=0.7, alpha=0.7)
    
    plt.legend()
    plt.show()

time1 = [0.004, 0.091, 3.84, 220.378, 17360.489]
time2 = [0.001, 0.007, 0.081, 1.012, 17.529]

time = [time1, time2]
plot_values(time)
```

```python
import matplotlib.pyplot as plt

# Tamanhos de entrada
entries = [1000, 10000, 100000, 1000000, 10000000]

# Tempos de execução estimados (em segundos) para Insertion Sort e Heap Sort
time_insertion_sort = [0.005, 0.05, 5, 500, 50000]
time_heap_sort = [0.001, 0.01, 0.1, 1, 10]

def plot_comparison(entries, time_insertion_sort, time_heap_sort):
    plt.figure(figsize=(10, 6))
    
    # Plotando Insertion Sort
    plt.plot(entries, time_insertion_sort, marker='o', color='blue', linestyle='-', label='Insertion Sort')
    plt.scatter(entries, time_insertion_sort, color='red', edgecolor='black')  # Destaque para os pontos
    
    # Plotando Heap Sort
    plt.plot(entries, time_heap_sort, marker='o', color='green', linestyle='-', label='Heap Sort')
    plt.scatter(entries, time_heap_sort, color='orange', edgecolor='black')  # Destaque para os pontos
    
    # Configurações do gráfico
    plt.xscale('log')  # Escala logarítmica para o eixo x
    plt.yscale('log')  # Escala logarítmica para o eixo y
    plt.xlabel('Tamanho das Entradas', fontsize=12)
    plt.ylabel('Tempo de Execução (s)', fontsize=12)
    plt.title('Comparação de Tempos de Execução: Insertion Sort vs Heap Sort', fontweight='bold', fontsize=14)
    plt.grid(True, which="both", ls="--", lw=0.5)
    plt.legend()
    plt.show()

# Chamando a função para plotar o gráfico
plot_comparison(entries, time_insertion_sort, time_heap_sort)
```
