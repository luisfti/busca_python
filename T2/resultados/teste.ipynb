```python
import matplotlib.pyplot as plt

def plot_values(time):
    entries = [1000, 10000, 100000, 1000000, 10000000]
    
    plt.figure(figsize=(10, 6))
    
    # Configurao do fundo branco
    plt.gcf().set_facecolor('white')  # Fundo branco para a figura
    
    # Plotando a primeira curva
    plt.plot(entries, time[0], marker='o', color='skyblue', linestyle='-', label='Curva utilizando Insertion Sort')
    
    # Plotando a segunda curva
    plt.plot(entries, time[1], marker='o', color='orange', linestyle='-', label='Curva utilizando Heap')
    
    plt.xscale('log')  # Escala logartmica para o eixo X
    plt.yscale('log')  # Escala logartmica para o eixo Y
    plt.xlabel('Tamanho das Entradas')
    plt.ylabel('Tempo (s)')
    plt.title('Grfico de Execuo dos Algoritmos', fontweight='bold')
    
    # Adicionando tracejados nos pontos
    for i in range(len(entries)):
        plt.axvline(x=entries[i], color='gray', linestyle='--', lw=0.7, alpha=0.7)
        plt.axhline(y=time[0][i], color='skyblue', linestyle='--', lw=0.7, alpha=0.7)
        plt.axhline(y=time[1][i], color='orange', linestyle='--', lw=0.7, alpha=0.7)
    
    plt.legend()
    plt.show()

time1 = [0.006, 0.088, 3.529, 204.843, 16481.107]
time2 = [0.003, 0.018, 1.301, 119.324, 13160.972]

time = [time1, time2]
plot_values(time)
```
