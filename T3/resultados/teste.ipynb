```python
import matplotlib.pyplot as plt

def plot_values(time):
    
    entries = [1000, 10000, 100000, 1000000]
    
    for i in range(len(time)):
        plt.figure(figsize=(10, 6))
        plt.plot(entries, time[i], marker='o', color='skyblue', linestyle='-')
        plt.xscale('log')  # Escala logartmica para melhor visualizao
        plt.xlabel('Nmero de Entradas')
        plt.ylabel('Tempo (s)')
        plt.title('Tempo de Execuo em Relao ao Nmero de Entradas')
        plt.grid(True, which="both", ls="--", lw=0.5)
        plt.text(1900000, 200, "Valores", fontsize=12, ha="left")
        plt.text(1900000, 180, time[i], fontsize=12, ha="left")
        plt.show()



time1 = [0.006, 0.098, 3.849, 218.631]
time2 = [0.003, 0.089, 3.501, 203.161]

time = [time1,time2]
plot_values(time)
```
