```python
import matplotlib.pyplot as plt

def plot_values(time):
    
    entries = [10000, 100000, 1000000, 10000000]
    
    for i in range(len(time)):
        plt.figure(figsize=(10, 6))
        plt.plot(entries, time[i], marker='o', color='skyblue', linestyle='-')
        plt.xscale('log')  # Escala logarítmica para melhor visualização
        plt.xlabel('Número de Entradas')
        plt.ylabel('Tempo (s)')
        if(i==0):
            plt.title('Tempo de Execução Sequencial')
        elif(i==1):
            plt.title('Tempo de Execução Sequencial Ordenado')
        else:
            plt.title('Tempo de Execução Binária')
        plt.grid(True, which="both", ls="--", lw=0.5)
        plt.show()



time1 = [0.0005, 0.0011, 0.0765, 0.0000]
time2 = [0.0003, 0.0025, 0.0919, 0.0000]
time3 = [0, 0, 0, 0]

time = [time1,time2, time3]
plot_values(time)
```
